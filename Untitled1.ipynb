```python
#!/usr/bin/env python3
"""
main.py
Detection + monocular depth fusion pipeline (single-file).

Features:
- YOLOv8 object detection (load once)
- Depth-Anything depth estimation (load once)
- Convert relative depth [0,1] -> approximate meters using DEPTH_MAX_METERS
- Annotated output image and CSV with detections + distances (meters & relative)
- Ablation study over confidence thresholds (reuses loaded models)
"""

# -------------------- 0. Imports --------------------
import os
import time
import csv
from typing import List, Dict
import cv2
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt
from ultralytics import YOLO
from transformers import AutoImageProcessor, AutoModelForDepthEstimation

# -------------------- 1. Config --------------------
IMG_PATH = "./leftImg8bit_trainvaltest/leftImg8bit/val/frankfurt/frankfurt_000000_003025_leftImg8bit.png"
OUT_DIR = "./outputs"
os.makedirs(OUT_DIR, exist_ok=True)

OUT_IMAGE = os.path.join(OUT_DIR, "annotated.png")
OUT_CSV = os.path.join(OUT_DIR, "detections_with_distance.csv")

YOLO_WEIGHTS = "yolov8n.pt"   # ensure this file is present or change to your path
DEPTH_MODEL_ID = "depth-anything/Depth-Anything-V2-small-hf"

# Convert relative depth to approximate meters by scaling:
# If predicted relative depth is 0 -> near (0 m), 1 -> far (DEPTH_MAX_METERS)
# Adjust DEPTH_MAX_METERS to match expected scene scale (e.g. 50 meters).
DEPTH_MAX_METERS = 50.0

# Obstacle classes to keep from COCO-like labels
OBSTACLE_CLASS_NAMES = {
    "person", "bicycle", "car", "motorcycle", "bus",
    "truck", "train", "traffic light", "stop sign", "bench"
}

# -------------------- 2. Helper Functions --------------------

def load_image(path: str) -> np.ndarray:
    if not os.path.isfile(path):
        raise FileNotFoundError(f"Image not found: {path}")
    img_bgr = cv2.imread(path, cv2.IMREAD_COLOR)
    if img_bgr is None:
        raise RuntimeError(f"Failed to load image as BGR: {path}")
    return img_bgr

def run_yolo(img_bgr: np.ndarray, model: YOLO, conf: float = 0.25):
    """
    Run YOLO on an image and return list of detections with format:
    {"cls_name": str, "conf": float, "xyxy": np.ndarray([x1, y1, x2, y2])}
    """
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    results = model.predict(source=img_rgb, conf=conf, verbose=False)[0]
    detections = []
    # model.names might be at model.model.names depending on Ultralytics version
    try:
        names = model.model.names
    except Exception:
        names = getattr(model, "names", {})
    for b in results.boxes:
        cls_id = int(b.cls.item())
        cls_name = names.get(cls_id, str(cls_id))
        if cls_name not in OBSTACLE_CLASS_NAMES:
            continue
        xyxy = b.xyxy.squeeze().cpu().numpy()
        conf_score = float(b.conf.item())
        detections.append({"cls_name": cls_name, "conf": conf_score, "xyxy": xyxy})
    return detections

def load_depth_anything(model_id: str = DEPTH_MODEL_ID):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    processor = AutoImageProcessor.from_pretrained(model_id)
    model = AutoModelForDepthEstimation.from_pretrained(model_id).to(device)
    model.eval()
    return processor, model, device

def run_depth_anything(img_bgr: np.ndarray, processor, model, device: str) -> np.ndarray:
    """
    Returns a normalized relative depth map in range [0,1] (float32).
    """
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    inputs = processor(images=img_rgb, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        predicted_depth = outputs.predicted_depth.squeeze().cpu().numpy().astype(np.float32)
    d_min, d_max = float(predicted_depth.min()), float(predicted_depth.max())
    # normalize to [0,1]
    rel = (predicted_depth - d_min) / (d_max - d_min + 1e-8)
    return rel.astype(np.float32)

def median_depth_in_box(depth_map: np.ndarray, xyxy: np.ndarray) -> float:
    x1, y1, x2, y2 = xyxy.astype(int)
    h, w = depth_map.shape[:2]
    x1, x2 = np.clip([x1, x2], 0, w - 1)
    y1, y2 = np.clip([y1, y2], 0, h - 1)
    if x2 <= x1 or y2 <= y1:
        return float("nan")
    roi = depth_map[y1:y2, x1:x2]
    return float(np.median(roi)) if roi.size > 0 else float("nan")

def direction_from_box(xyxy: np.ndarray, img_w: int) -> str:
    cx = 0.5 * (xyxy[0] + xyxy[2])
    if cx < img_w / 3:
        return "Left"
    elif cx > 2 * img_w / 3:
        return "Right"
    return "Center"

def rel_to_meters(rel_depth: float, depth_max_m: float = DEPTH_MAX_METERS) -> float:
    """Convert normalized relative depth [0,1] to approximate meters."""
    if np.isnan(rel_depth):
        return float("nan")
    return float(rel_depth * depth_max_m)

def annotate_and_save(img_bgr: np.ndarray,
                      detections: List[Dict],
                      depth_map: np.ndarray,
                      out_image_path: str,
                      out_csv_path: str,
                      depth_max_m: float = DEPTH_MAX_METERS):
    """
    Annotate image (draw boxes + labels) and save CSV with detection info.
    CSV columns: class, confidence, distance_rel, distance_m, direction, x1,y1,x2,y2
    """
    h, w = img_bgr.shape[:2]
    vis = img_bgr.copy()

    # Sort by median depth (near -> far). Handle NaN by placing them at the end.
    def _depth_for_sort(d):
        val = median_depth_in_box(depth_map, d["xyxy"])
        return val if not np.isnan(val) else 1.0  # push NaN to far end

    detections_sorted = sorted(detections, key=_depth_for_sort)

    with open(out_csv_path, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["class", "confidence", "distance_rel", "distance_m", "direction", "x1", "y1", "x2", "y2"])
        for det in detections_sorted:
            dist_rel = median_depth_in_box(depth_map, det["xyxy"])
            dist_m = rel_to_meters(dist_rel, depth_max_m)
            direction = direction_from_box(det["xyxy"], w)
            x1, y1, x2, y2 = det["xyxy"].astype(int)

            # Draw bounding box
            cv2.rectangle(vis, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Prepare label text
            label = f"{det['cls_name']} {det['conf']:.2f} | {dist_m:.1f}m"
            # Draw black outline for readability, then white text
            cv2.putText(vis, label, (x1, max(12, y1 - 6)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), thickness=3, lineType=cv2.LINE_AA)
            cv2.putText(vis, label, (x1, max(12, y1 - 6)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), thickness=1, lineType=cv2.LINE_AA)

            writer.writerow([
                det["cls_name"],
                round(det["conf"], 4),
                round(float(dist_rel) if not np.isnan(dist_rel) else float("nan"), 4),
                round(float(dist_m) if not np.isnan(dist_m) else float("nan"), 3),
                direction,
                int(x1), int(y1), int(x2), int(y2)
            ])

    cv2.imwrite(out_image_path, vis)
    return vis

# -------------------- 3. Main Pipeline --------------------

def main():
    print("[INFO] Starting detection + depth fusion pipeline")
    total_start = time.time()

    # Load image
    img_bgr = load_image(IMG_PATH)
    h, w = img_bgr.shape[:2]
    print(f"[INFO] Loaded image {IMG_PATH} (w={w}, h={h})")

    # Load models once
    t0 = time.time()
    print("[INFO] Loading YOLO model...")
    yolo_model = YOLO(YOLO_WEIGHTS)
    print(f"[INFO] Loading depth model ({DEPTH_MODEL_ID})...")
    processor, depth_model, device = load_depth_anything(DEPTH_MODEL_ID)
    t1 = time.time()
    print(f"[INFO] Models loaded in {t1 - t0:.2f}s (device={device})")

    # Run YOLO at default confidence
    conf_default = 0.25
    t0 = time.time()
    detections = run_yolo(img_bgr, model=yolo_model, conf=conf_default)
    t1 = time.time()
    print(f"[INFO] YOLO inference ({len(detections)} detections) took {t1 - t0:.2f}s")

    if len(detections) == 0:
        print("[WARN] No detections found. Creating an empty CSV and saving input image copy.")
        # Save empty CSV
        with open(OUT_CSV, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["class", "confidence", "distance_rel", "distance_m", "direction", "x1", "y1", "x2", "y2"])
        # Save copy of the input image
        cv2.imwrite(OUT_IMAGE, img_bgr)
        return

    # Run depth model (relative depth map)
    t0 = time.time()
    depth_rel = run_depth_anything(img_bgr, processor, depth_model, device)
    t1 = time.time()
    print(f"[INFO] Depth estimation took {t1 - t0:.2f}s")

    # Annotate & save main outputs
    annotated = annotate_and_save(img_bgr, detections, depth_rel, OUT_IMAGE, OUT_CSV, DEPTH_MAX_METERS)
    print(f"[INFO] Saved annotated image -> {OUT_IMAGE}")
    print(f"[INFO] Saved detections CSV -> {OUT_CSV}")

    # Quantitative summary
    df = pd.read_csv(OUT_CSV)
    print("\n--- Quantitative Analysis ---")
    print("Total objects detected:", len(df))
    if len(df) > 0:
        print("Mean distance (m):", df["distance_m"].mean())
        print("Closest obstacle:\n", df.loc[df["distance_m"].idxmin()])
        print("Farthest obstacle:\n", df.loc[df["distance_m"].idxmax()])
        print("\nDetections per class:\n", df["class"].value_counts())
        print("\nDirection distribution:\n", df["direction"].value_counts())

    # Quick visualization (optional)
    try:
        plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
        plt.title("Input Image")
        plt.axis("off")
        plt.show()

        plt.imshow(depth_rel, cmap="inferno")
        plt.colorbar(label="Relative Depth (0=near, 1=far)")
        plt.title("Predicted Relative Depth Map")
        plt.axis("off")
        plt.show()

        plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
        plt.title("Annotated Detections (YOLO + Depth)")
        plt.axis("off")
        plt.show()
    except Exception:
        # If running headless, skip plotting
        pass

    # Save some plots to disk for analysis (confidence vs distance)
    if len(df) > 0:
        plt.figure(figsize=(8, 5))
        plt.scatter(df["confidence"], df["distance_m"], alpha=0.7)
        plt.xlabel("YOLO Confidence")
        plt.ylabel("Distance (m)")
        plt.title("Confidence vs Distance (m)")
        plt.grid(True)
        plt.savefig(os.path.join(OUT_DIR, "confidence_vs_distance.png"))
        plt.close()

        plt.figure(figsize=(10, 5))
        df.groupby("class")["distance_m"].mean().sort_values().plot(kind="bar")
        plt.ylabel("Mean Distance (m)")
        plt.title("Mean Distance by Object Class")
        plt.tight_layout()
        plt.savefig(os.path.join(OUT_DIR, "mean_distance_per_class.png"))
        plt.close()

    # -------------------- Ablation: Confidence Thresholds --------------------
    for conf in [0.25, 0.5, 0.75]:
        print(f"\n[ABLA] Running ablation with confidence={conf}")
        detections_conf = run_yolo(img_bgr, model=yolo_model, conf=conf)
        out_csv_conf = OUT_CSV.replace(".csv", f"_{int(conf*100)}.csv")
        out_img_conf = OUT_IMAGE.replace(".png", f"_{int(conf*100)}.png")
        if not detections_conf:
            print(f"[ABLA] No detections at conf={conf}, writing empty CSV.")
            with open(out_csv_conf, "w", newline="") as f:
                writer = csv.writer(f)
                writer.writerow(["class", "confidence", "distance_rel", "distance_m", "direction", "x1", "y1", "x2", "y2"])
            cv2.imwrite(out_img_conf, img_bgr)
            continue
        annotate_and_save(img_bgr, detections_conf, depth_rel, out_img_conf, out_csv_conf, DEPTH_MAX_METERS)
        df_conf = pd.read_csv(out_csv_conf)
        print(f"[ABLA] conf={conf} -> total detections: {len(df_conf)}, mean dist (m): {df_conf['distance_m'].mean():.2f}")

    total_end = time.time()
    print(f"\n[INFO] Total pipeline time: {total_end - total_start:.2f} seconds")

if __name__ == "__main__":
    main()
```

```python
# -------------------- 1. Imports --------------------
import os, time, csv
import cv2
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt
from ultralytics import YOLO
from transformers import AutoImageProcessor, AutoModelForDepthEstimation

# -------------------- 2. Config --------------------
IMG_PATH = "./leftImg8bit_trainvaltest/leftImg8bit/val/frankfurt/frankfurt_000000_003025_leftImg8bit.png"
OUT_DIR = "./outputs_yolo_versions"
os.makedirs(OUT_DIR, exist_ok=True)

OBSTACLE_CLASS_NAMES = {
    "person", "bicycle", "car", "motorcycle", "bus", "truck", "train",
    "traffic light", "stop sign", "bench"
}

YOLO_VERSIONS = ["yolov8n.pt", "yolov8s.pt", "yolov8m.pt", "yolov8l.pt"]

# -------------------- 3. Helper Functions --------------------
def load_image(path: str) -> np.ndarray:
    if not os.path.isfile(path):
        raise FileNotFoundError(f"Image not found: {path}")
    return cv2.imread(path, cv2.IMREAD_COLOR)

def run_yolo(img_bgr: np.ndarray, weights: str, conf: float = 0.25):
    model = YOLO(weights)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    results = model.predict(source=img_rgb, conf=conf, verbose=False)[0]
    detections = []
    names = model.model.names
    for b in results.boxes:
        cls_id = int(b.cls.item())
        cls_name = names.get(cls_id, str(cls_id))
        if cls_name not in OBSTACLE_CLASS_NAMES:
            continue
        xyxy = b.xyxy.squeeze().cpu().numpy()
        conf_score = float(b.conf.item())
        detections.append({"cls_name": cls_name, "conf": conf_score, "xyxy": xyxy})
    return detections

def load_depth_anything(model_id: str = "depth-anything/Depth-Anything-V2-small-hf"):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    processor = AutoImageProcessor.from_pretrained(model_id)
    model = AutoModelForDepthEstimation.from_pretrained(model_id).to(device)
    model.eval()
    return processor, model, device

def run_depth_anything(img_bgr: np.ndarray, processor, model, device: str) -> np.ndarray:
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    inputs = processor(images=img_rgb, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        predicted_depth = outputs.predicted_depth.squeeze().cpu().numpy().astype(np.float32)
    d_min, d_max = float(predicted_depth.min()), float(predicted_depth.max())
    return (predicted_depth - d_min) / (d_max - d_min + 1e-8)

def median_depth_in_box(depth_map: np.ndarray, xyxy: np.ndarray) -> float:
    x1, y1, x2, y2 = xyxy.astype(int)
    h, w = depth_map.shape[:2]
    x1, x2 = np.clip([x1, x2], 0, w - 1)
    y1, y2 = np.clip([y1, y2], 0, h - 1)
    if x2 <= x1 or y2 <= y1:
        return float("nan")
    roi = depth_map[y1:y2, x1:x2]
    return float(np.median(roi)) if roi.size > 0 else float("nan")

def direction_from_box(xyxy: np.ndarray, img_w: int) -> str:
    cx = 0.5 * (xyxy[0] + xyxy[2])
    if cx < img_w / 3:
        return "Left"
    elif cx > 2 * img_w / 3:
        return "Right"
    return "Center"

def annotate_and_save(img_bgr, detections, depth_map, out_image_path, out_csv_path):
    h, w = img_bgr.shape[:2]
    vis = img_bgr.copy()
    with open(out_csv_path, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["class", "confidence", "distance_rel", "direction", "x1", "y1", "x2", "y2"])
        detections_sorted = sorted(detections, key=lambda d: median_depth_in_box(depth_map, d["xyxy"]))
        for det in detections_sorted:
            dist_rel = median_depth_in_box(depth_map, det["xyxy"])
            direction = direction_from_box(det["xyxy"], w)
            x1, y1, x2, y2 = det["xyxy"].astype(int)
            cv2.rectangle(vis, (x1, y1), (x2, y2), (0, 255, 0), 2)
            label = f"{det['cls_name']} {det['conf']:.2f} | rel={dist_rel:.2f} | {direction}"
            cv2.putText(vis, label, (x1, max(0, y1 - 5)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
            writer.writerow([det["cls_name"], f"{det['conf']:.4f}", f"{dist_rel:.4f}", direction, x1, y1, x2, y2])
    cv2.imwrite(out_image_path, vis)
    return vis

# -------------------- 4. Main Experiment: Multiple YOLO Versions --------------------
img_bgr = load_image(IMG_PATH)
processor, depth_model, device = load_depth_anything()
depth_rel = run_depth_anything(img_bgr, processor, depth_model, device)

results_summary = []

for weights in YOLO_VERSIONS:
    print(f"\n[INFO] Running pipeline with {weights}...")
    start = time.time()

    detections = run_yolo(img_bgr, weights=weights, conf=0.25)
    out_img = os.path.join(OUT_DIR, f"annotated_{weights.replace('.pt', '')}.png")
    out_csv = os.path.join(OUT_DIR, f"detections_{weights.replace('.pt', '')}.csv")
    annotate_and_save(img_bgr, detections, depth_rel, out_img, out_csv)

    df = pd.read_csv(out_csv)
    elapsed = time.time() - start

    results_summary.append({
        "model": weights,
        "num_detections": len(df),
        "mean_depth": df["distance_rel"].mean(),
        "min_depth": df["distance_rel"].min(),
        "inference_time_s": round(elapsed, 3)
    })

# -------------------- 5. Summary Table --------------------
summary_df = pd.DataFrame(results_summary)
print("\n=== YOLO Model Comparison ===")
print(summary_df)

# Plot detections vs. model size
plt.figure(figsize=(8,5))
plt.bar(summary_df["model"], summary_df["num_detections"], color="teal")
plt.ylabel("Total Detections")
plt.title("Detections by YOLO Version")
plt.grid(axis="y")
plt.show()

# Plot inference time comparison
plt.figure(figsize=(8,5))
plt.bar(summary_df["model"], summary_df["inference_time_s"], color="orange")
plt.ylabel("Inference Time (s)")
plt.title("Inference Time by YOLO Version")
plt.grid(axis="y")
plt.show()
```
